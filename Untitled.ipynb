```python
# Identify columns containing infrastructure scores
infrastructure_columns = [col for col in infrastructure_quality_data.columns if 'decimal' in col]

# Convert columns to numeric
infrastructure_quality_data[infrastructure_columns] = infrastructure_quality_data[infrastructure_columns].apply(pd.to_numeric, errors='coerce')

# Calculate percentage change between each pair of consecutive years
percentage_changes = pd.DataFrame(index=infrastructure_quality_data.index)

for i in range(len(infrastructure_columns) - 1):
    year_1 = infrastructure_columns[i]
    year_2 = infrastructure_columns[i + 1]
    percentage_changes[f'Change from {year_1} to {year_2}'] = (
        (infrastructure_quality_data[year_2] - infrastructure_quality_data[year_1]) /
        infrastructure_quality_data[year_1] * 100
    )

# Add these percentage changes to the existing DataFrame
infrastructure_quality_data = pd.concat([infrastructure_quality_data, percentage_changes], axis=1)

# Save the updated DataFrame to the existing file
infrastructure_quality_data.to_excel('infrastructure quality.xlsx', index=False)

print("Detailed percentage changes in Infrastructure Quality have been saved to the existing file.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
mortality_data = pd.read_excel('mortality.xlsx')
investment_data = pd.read_excel('infrastructure spending per country.xlsx')

# Define the years for plotting
years = ['1990 [decimal1990]', '2000 [decimal2000]', '2014 [decimal2014]', '2015 [decimal2015]', '2016 [decimal2016]', '2017 [decimal2017]', '2018 [decimal2018]', '2019 [decimal2019]', '2020 [decimal2020]', '2021 [decimal2021]', '2022 [decimal2022]', '2023 [decimal2023]']

# Filter and aggregate data for each year
mortality_avg = mortality_data[years].mean()
investment_avg = investment_data[years].mean()

# Ensure the data matches the years for plotting
mortality_avg = mortality_avg[years]
investment_avg = investment_avg[years]

# Define the years for plotting
years_plot = [year.split('[')[0].strip() for year in years]

# Descriptive statistics
print("Descriptive Statistics for Average Annual Mortality Rate:")
print(mortality_avg.describe())
print("\nDescriptive Statistics for Average Annual Investment:")
print(investment_avg.describe())

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(years_plot, mortality_avg, label='Average Annual Mortality Rate', marker='o', color='red')
plt.plot(years_plot, investment_avg, label='Average Annual Investment', marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Comparison of Average Annual Mortality Rate and Average Annual Investment Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('comparison_mortality_investment.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
infrastructure_data = pd.read_excel('infrastructure quality.xlsx')
investment_data = pd.read_excel('infrastructure spending per country.xlsx')
mortality_data = pd.read_excel('mortality.xlsx')

# Define years
years = ['1990 [decimal1990]', '2000 [decimal2000]', '2014 [decimal2014]', '2015 [decimal2015]', '2016 [decimal2016]', '2017 [decimal2017]', '2018 [decimal2018]', '2019 [decimal2019]', '2020 [decimal2020]', '2021 [decimal2021]', '2022 [decimal2022]', '2023 [decimal2023]']

# Calculate average values for each year
avg_quality_scores = infrastructure_data[years].mean()
avg_investment_levels = investment_data[years].mean()
avg_mortality_rates = mortality_data[years].mean()

# Calculate percentage change for each
percentage_change_quality = avg_quality_scores.pct_change() * 100
percentage_change_investment = avg_investment_levels.pct_change() * 100
percentage_change_mortality = avg_mortality_rates.pct_change() * 100

# Descriptive statistics
stats_quality = percentage_change_quality.describe()
stats_investment = percentage_change_investment.describe()
stats_mortality = percentage_change_mortality.describe()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(percentage_change_quality, label='Percentage Change in Quality Score', marker='o', color='red')
plt.plot(percentage_change_investment, label='Percentage Change in Investment Levels', marker='o', color='blue')
plt.plot(percentage_change_mortality, label='Percentage Change in Mortality Rate', marker='o', color='green')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.title('Comparison of Percentage Change Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plot
plt.savefig('comparison_percentage_change.png')
plt.show()

# Print descriptive statistics
print("Descriptive Statistics for Percentage Change in Quality Score:")
print(stats_quality)
print("\nDescriptive Statistics for Percentage Change in Investment Levels:")
print(stats_investment)
print("\nDescriptive Statistics for Percentage Change in Mortality Rate:")
print(stats_mortality)
```

```python
import pandas as pd

# Load the data
mortality_data = pd.read_excel('mortality.xlsx')
infrastructure_data = pd.read_excel('infrastructure quality.xlsx')
investment_data = pd.read_excel('infrastructure spending per country.xlsx')

# Extract columns containing 'decimal'
mortality_columns = [col for col in mortality_data.columns if 'decimal' in col]
infrastructure_columns = [col for col in infrastructure_data.columns if 'decimal' in col]
investment_columns = [col for col in investment_data.columns if 'decimal' in col]

# Ensure the columns are numeric
mortality_data[mortality_columns] = mortality_data[mortality_columns].apply(pd.to_numeric, errors='coerce')
infrastructure_data[infrastructure_columns] = infrastructure_data[infrastructure_columns].apply(pd.to_numeric, errors='coerce')
investment_data[investment_columns] = investment_data[investment_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the Average Annual Mortality Rate for each country
mortality_data['Average Annual Mortality Rate'] = mortality_data[mortality_columns].mean(axis=1)

# Calculate the Average Infrastructure Quality Score for each country
infrastructure_data['Average Infrastructure Quality Score'] = infrastructure_data[infrastructure_columns].mean(axis=1)

# Calculate the Average Annual Investment for each country
investment_data['Average Annual Investment'] = investment_data[investment_columns].mean(axis=1)

# Print summary statistics for all relevant columns
print("Summary Statistics for Mortality Data:")
print(mortality_data.describe(include='all'))
print("\nSummary Statistics for Infrastructure Data:")
print(infrastructure_data.describe(include='all'))
print("\nSummary Statistics for Investment Data:")
print(investment_data.describe(include='all'))

# Descriptive statistics for calculated averages
print("\nDescriptive Statistics for Average Annual Mortality Rate:")
print(mortality_data['Average Annual Mortality Rate'].describe())
print("\nDescriptive Statistics for Average Infrastructure Quality Score:")
print(infrastructure_data['Average Infrastructure Quality Score'].describe())
print("\nDescriptive Statistics for Average Annual Investment:")
print(investment_data['Average Annual Investment'].describe())
```
